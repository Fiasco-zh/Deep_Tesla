```python
#引入所需库文件

import cv2
import numpy as np
import pandas as pd
import pickle
```

```python
#设定数据路径

mkv_file_path = ['../epochs/epoch{:02d}_front.mkv'.format(i) for i in range(1,11)]
steering_file_path = ['../epochs/epoch{:02d}_steering.csv'.format(i) for i in range(1,11)]
paths = [mkv_file_path,steering_file_path]
```

```python
#数据预处理方法

def read_mkv(path):
    cap = cv2.VideoCapture(path)
    success,frame = cap.read()
    pic = []

    while(success):
        frame = frame[300:-50]
        frame = cv2.resize(frame,(224,64),interpolation=cv2.INTER_CUBIC)
        pic.append(frame)

        success,frame = cap.read()
    return np.array(pic)

def read_steering(path):
    return pd.read_csv(path).wheel.values
```

```python
# 读取epoch1~8视频和转向角度,作为训练集
train_x=[]
train_y = []
for i in range(0,8):
    train_x.append(read_mkv(paths[0][i]))
    train_y.append(read_steering(paths[1][i]))

#读取epoch9，作为验证集
val_x = read_mkv(paths[0][8])
val_y = read_steering(paths[1][8])
```

```python
# 保存视频和转向角度

train_x = np.concatenate(tuple(train_x))
train_y = np.concatenate(tuple(train_y))
pickle.dump((train_x,train_y,val_x,val_y),open('../data/preprocess.p','wb'))
```

```python
# 加载数据

train_x,train_y,val_x,val_y = pickle.load(open('../data/preprocess.p','rb'))
print('data load successfully')
print('train_x shape is {}'.format(train_x.shape))
```

```python
#查看处理后的图片

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
plt.imshow(train_x[np.random.randint(0,train_x.shape[0])])
```

```python
#翻转图像和对应转向角，增加一倍训练数据，目前来看对结果没有提升效果

reversed_x = np.array([cv2.flip(image,1) for image in train_x])
reversed_y = -1*train_y
```

```python
feature = np.concatenate((train_x,reversed_x))
label = np.concatenate((train_y,reversed_y))
print('feature shape is {}'.format(feature.shape))
```

```python
#导入keras相关库文件

from keras.layers import Dense,Activation,Convolution2D,MaxPooling2D,Flatten,Dropout,BatchNormalization
from keras.layers.noise import GaussianNoise
from keras.models import Sequential
from keras import regularizers
```

## My Net

```python
#构建模型

model = Sequential()

model.add(BatchNormalization(input_shape=(64,224,3))) #对数据进行归一化 
# model.add(GaussianNoise(5,input_shape=(66,228,3)))  #对数据增加噪音，经测试没有提升结果
model.add(Convolution2D(kernel_size=(5,5),filters = 8,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))

# model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(BatchNormalization())

model.add(Convolution2D(kernel_size=(3,3),filters = 64,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(BatchNormalization())

model.add(Convolution2D(kernel_size=(3,3),filters = 64,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
# model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Activation('relu'))
# model.add(BatchNormalization())

model.add(Convolution2D(kernel_size=(3,3),filters = 128,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))

# model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(BatchNormalization())

model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
# model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

# model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),data_format='channels_last'))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),data_format='channels_last'))
# model.add(Activation('relu'))
# # model.add(BatchNormalization())

# model.add(Convolution2D(kernel_size=(3,3),filters = 256,strides=(1,1),data_format='channels_last'))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

# model.add(Dense(512))
# model.add(Activation('relu'))
# # model.add(BatchNormalization())
# model.add(Dropout(0.75))

model.add(Dense(1024,activation='relu',kernel_regularizer=regularizers.l2(),bias_regularizer=regularizers.l2()))
# model.add(BatchNormalization())
model.add(Dropout(0.5))

# model.add(Dense(1024))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))

model.add(Dense(1))
```

```python
#自定义评估函数，33是max(steering)-min(steering)

import keras.backend as K

def missing_rate(y_true, y_pred):
    return K.mean(abs(y_pred-y_true))/33.
```

```python
model.compile(optimizer='adadelta',
             loss = 'mean_squared_error',
             metrics = [missing_rate])
```

```python
# 模型概览

model.summary()
```

```python
#引入图片生成器，并做了通道随机偏移，也无法提升结果，故取消了，目前仅保留生成batch数据的功能

from keras.preprocessing.image import ImageDataGenerator
generator = ImageDataGenerator(data_format='channels_last')
```

```python
#训练模型,可多次运行，连续训练模型

model.fit_generator(generator.flow(feature,label,batch_size=64),steps_per_epoch=len(feature)/64,\
                    epochs=3,validation_data=(val_x,val_y))
```

```python
# generator在反复训练时会出现内存不足问题，故又采用fit()方法

model.fit(feature,label,batch_size=64,shuffle=True,validation_data=(val_x,val_y),epochs=5)
```

```python
#保存模型配置和权重

import h5py
model.save_weights('../models/model_0708.h5')
with open('../models/model_0708.json','w') as f:
    f.write(model.to_json())
    f.close()
```
